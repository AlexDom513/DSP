```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy import fftpack
```

```python
img = cv2.imread('test2.jpg', 0)
#img = cv2.imread('test2.jpg', 0)

imgDisp = cv2.resize(img, (0,0), fx=0.25, fy=0.25)
print(img.shape)

cv2.imshow('image', imgDisp)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.imshow(img)
```

```python
blur = cv2.blur(img, (80,80))
plt.imshow(blur)
```

```python
#contrast measurement
image = img
height, width = image.shape[:2]
def px(x, y):
    return int(image[y, x])

sum = 0
for x in range(width-1):
    for y in range(height):
        sum += abs(px(x, y) - px(x+1, y))
print(sum)
```

```python
#variance of the laplacian
image = img
fm = cv2.Laplacian(image, cv2.CV_64F).var()
plt.imshow(image)
print(fm)
```

```python
image = blur
fm = cv2.Laplacian(image, cv2.CV_64F).var()
plt.imshow(image)
print(fm)
```
