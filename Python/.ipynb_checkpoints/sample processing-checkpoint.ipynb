```python
import numpy as np
import scipy.signal as sig
import matplotlib.pyplot as plt

fs = 1000
freq = 10
omega = 2 * np.pi * freq
print(str(omega)+' Rad/s')

x = np.arange(0,1,1/fs)
sine = np.sin(omega*x)
plt.plot(x,sine)
plt.show()

numtaps = 40
cutoff = 5
filt = sig.firwin(numtaps, cutoff, fs=fs)
#need to use fs

signal = np.convolve(sine,filt)
plt.plot(sine)
plt.show()
plt.plot(signal,color='orange')
plt.show()

w,h = sig.freqz(filt,1)
w = (w * fs) / (2*np.pi)
plt.plot(w,20 * np.log10(abs(h)))
plt.axvline(cutoff, color='green')
```
