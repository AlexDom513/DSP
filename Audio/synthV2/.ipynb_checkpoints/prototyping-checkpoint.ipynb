```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack
from scipy import signal
from scipy.io.wavfile import write
import IPython
```

# General

```python
fs = 500
freq = 10

start = 0
stop = 1
numSamples = (stop-start)*fs

x = np.linspace(start, stop, numSamples)
sine = np.sin(2*np.pi*x*freq)
plt.plot(sine)
```

part1: the comb filter

```python
weights = [1,.8]
w, h = signal.freqz(weights, whole=True)
plt.plot(w/(2*np.pi), np.abs(h))
plt.xlabel('Normalized Frequency')
plt.ylabel('Magnitude')
plt.title('Digital Comb Filter ({} tap delay)'.format(len(weights)-1))
plt.tight_layout()
```

```python
#difference eqn: y[n] = x[n] + αx[n-k]
#impulse response: h[n] = δ[n] + α(δ[n-k])

def combFilt(signal, k, g):
    return signal[:-k] + g*signal[k:]

#condition: k2>k1
g = .2
k1 = 1
k2 = 25
filtSig1 = combFilt(sine, k1, g)
filtSig2 = combFilt(sine, k2, g)

#add all the waveforms together
combCombined = sine[:-k2]+filtSig1[:-abs(k1-k2)]+filtSig2

plt.figure(figsize=(20,10))
plt.plot(x,sine)
plt.plot(x[k1:],filtSig1)
plt.plot(x[k2:],filtSig2)
plt.plot(x[k2:],combCombined)
plt.axhline(y=0, color='r')
plt.legend(['sine','filtSig1','filtSig2','combCombined'])

#write file for listening
write('combCombined.wav', fs, combCombined.astype(np.float32))
write('sine.wav', fs, sine.astype(np.float32))
```
