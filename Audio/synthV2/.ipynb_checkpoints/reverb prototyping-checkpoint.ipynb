Reverb: an effect which creates a sense of space, by generating a diffuse longer-lasting sound from a short input<br><br>
Methods:<br>convolution (complicated, high CPU usage, but can imitate real spaces well),<br>
feedback delay network (very efficient, tricky to make realistic)

```python
#impulse response
samplerate, x = read('impulseResponse.wav')
plt.plot(x)
print(str(samplerate) + ' Hz')
IPython.display.Audio('impulseResponse.wav')
```

```python
#comb filter
y = np.zeros(len(x))
gains = [.95, .85, .5, .005]
delay = 1

#changing gain parameter
for gain in gains:
    for i in range(len(x)):
        if i >= delay:
            y[i] = x[i] + gain*y[i-delay]
    print(gain)
    plt.plot(y)
    plt.show()
```

```python
#changing delay parameter
y = np.zeros(len(x))
gain = .8
delays = [1, 5, 10, 15, 20]

for delay in delays:
    for i in range(len(x)):
        if i >= delay:
            y[i] = x[i] + gain*y[i-delay]
    print(delay)
    plt.plot(y)
    plt.show()
```

```python
#listen to effect of delay paramter
y = np.zeros(len(x))
gain = .8
delay = 20

for i in range(len(x)):
    if i >= delay:
        y[i] = x[i] + gain*y[i-delay]
        
print(delay)
plt.plot(y)
write('combedImpulse.wav', samplerate, y.astype(np.int16))
IPython.display.Audio('combedImpulse.wav')
```

```python
#apply effect to snare
samplerate, x = read('cantina.wav')
plt.plot(x)
IPython.display.Audio('cantina.wav')
```

```python
#apply effect to snare
y = np.zeros(len(x))
gain = .6
delay1 = 10
delay2 = 20
delay3 = 30

for i in range(len(x)):
    if i >= delay3:
        y[i] = x[i] + gain*y[i-delay1] - gain*y[i-delay2] + gain*y[i-delay3]

plt.plot(y)
write('combedCantina.wav', samplerate, y.astype(np.int16))
IPython.display.Audio('combedCantina.wav')
```
